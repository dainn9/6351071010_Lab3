```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
%matplotlib inline

STUDENT = 'Nguyễn Trần Thanh Danh - 6351071010'
```

```python
from PIL import Image, ImageOps
import requests
from io import BytesIO
url = 'https://raw.githubusercontent.com/dainn9/6351071010_Lab3/master/dog.jpg'
response = requests.get(url)
response.raise_for_status()
img = Image.open(BytesIO(response.content))
axes = plt.axes(xticks=[], yticks=[])
print(STUDENT)
axes.imshow(img)
```

```python
imarr = np.array(img)
rows = imarr.shape[0]
cols = imarr.shape[1]

data = imarr/255.0
data = data.reshape(rows * cols, 3)
print(STUDENT)
data.shape
```

```python
kmeans = KMeans(n_clusters = 32, max_iter=300)
kmeans.fit(data)
new_colors = kmeans.cluster_centers_[kmeans.predict(data)]

img_recolored = new_colors.reshape(imarr.shape)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16,6), subplot_kw=dict(xticks=[],yticks=[]))

print(STUDENT)

fig.subplots_adjust(wspace=0.05)
ax[0].imshow(img)
ax[0].set_title('Original Image', size=16)
ax[1].imshow(img_recolored)
ax[1].set_title('32-color Image', size=16);
```
